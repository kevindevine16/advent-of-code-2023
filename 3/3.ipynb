```python
def getNumbersOnly(row):
    nums = []
    start_indexes = []
    end_indexes = []
    num_lengths = []
    j = 0
    while j < len(row):
        k = 1
        if row[-1-j].isdigit():
            digit = True 
            while digit:
                if j == len(row)-1:
                    digit = False
                elif not row[-1-j-k].isdigit():
                    digit = False
                else:
                    k  +=1
                    if k+j == len(a):
                        digit = False
            num = 0
            for kk in range(k):
                num += int(row[-1-j-kk])*(10**kk)
            nums.append(num)
            start_indexes.append(len(a)-j-k)
            end_indexes.append(len(a)-1-j)
            num_lengths.append(k)
            j = j+k

        j +=1    
    return nums, num_lengths, start_indexes, end_indexes
```

```python
def FindGears(a):
    num_gears = 0
    Gears_indices = []
    for i in range(len(a)):
        for j in range(len(a[i])):
            if a[i][j] == '*':
                Gears_indices.append([i,j])
                num_gears += 1
    print(num_gears)
    return Gears_indices

Gears_Indices = FindGears(a)
```
